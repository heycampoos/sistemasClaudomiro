
5.2 Considere um sistema de computação com uma memória cache de 32KB de capacidade, constituída de linhas com 8 bytes de largura. A MP possui uma capacidade de 16MB.

 _Aqui, uma função que realiza os calculos após a redução para a base 2 seria necessária para transformar a resolução em um código Python_.

```python
# Esta função tem como utilidade a transformação do valor em bits
def base2Redutor(valor, unidade):
    convert = 0
    if unidade.lower() == 'k':
        convert = valor * 1024
    if unidade.lower() == 'b':
        convert = valor
    if unidade.lower() == 'm':
        convert = valor * 1024**2
    if unidade.lower() == 'kb':
        convert = valor * 2 ** 10
    if unidade.lower() == 'mb':
        convert = valor * 2 ** 20
 
    for i in range(0, 255):
            if 2 ** i >= convert: return i
#Aqui, seguindo, temos a calculadora de bits que irá efetuar os cálculos
def calcBits(quantiCelulas, bitsCelula, cap_memPrincipal):
    totBitsDados = quantiCelulas * bitsCelula * 1024
    '''Acima, o total de bits para a parte de dados. Abaixo, a quantidade de linhas e blocos da cache'''
    linhas = base2Redutor(quantiCelulas / bitsCelula, "kb")
    blocos = base2Redutor(quantiCelulas / bitsCelula, "mb")
    '''Para a largura do campo de Tag: e o total de bits'''
    blocoLinha_tag = blocos - linhas 
    totTags = 2 ** linhas * blocoLinha_tag
    '''Total geral de bits'''
    totalBits = totBitsDados + totTags
    
    return totalBits

    
print(f'''Para {calcBits(32, 8, 16)}, nós teremos:
{calcBits(32, 8, 16) / 1024}KB''')
```


5.4 Seja uma MP constituída de blocos com largura de 32 bytes, associada a uma cache com 128KB. Em dado instante o processador realiza um acesso, colocando o seguinte endereço (expresso em algarismos hexadecimais): 3FC92B6. Determine qual deverá ser o valor binário da linha que será localizada pelo sistema de controle da cache. 

_Aqui, teremos um bloco responsável pela conversão do valor hexadecimal parra binário da seguinte forma:_

```python
  
def dec_Bin(a):
    number = int(a)
    string = ''
    while number > 0:
        rest = int(number % 2)
        string += str(rest)
        number = (number - rest) / 2
    return string[::-1]

def Hex_Bin(v):
    return len(v) * 4

def enderecoLinhaBits(larg_bloco, cap_memPrincipal, h):
    EnderecoHex = Hex_Bin(h) 
    CampoByte = base2Redutor(larg_bloco, 'b') 
    CampoLinha = base2Redutor(cap_memPrincipal / larg_bloco, 'kb') 
    CampoTag = EnderecoHex - (CampoByte + CampoLinha) 
    enderecoBinario = dec_Bin(int(h, 16))
    
    # Correção zero
    if len(enderecoBinario) < EnderecoHex:
        while len(enderecoBinario) < EnderecoHex:
            temp = enderecoBinario[::-1]
            temp += '0'
            enderecoBinario = temp[::-1]
    
    return enderecoBinario[CampoTag:(CampoTag+CampoLinha)]
    
print(f'Endereço da linha desejada: {enderecoLinhaBits(32, 128, "3FC92B6")}')
```

5.5 Cálculo da quantidade de bits necessários para uma determinada memória cache. 
Considere um sistema de computação com uma memória cache de 32KB de capacidade, constituída de linhas com 8 bytes de largura. A MP possui uma capacidade de 16MB. 

```python
def base2Redutor(valor, unidade):
    convert = 0
    if unidade.lower() == 'k':
        convert = valor * 1024
    if unidade.lower() == 'b':
        convert = valor
    if unidade.lower() == 'm':
        convert = valor * 1024**2
    if unidade.lower() == 'kb':
        convert = valor * 2 ** 10
    if unidade.lower() == 'mb':
        convert = valor * 2 ** 20
 
    for i in range(0, 255):
            if 2 ** i >= convert: return i
            
def quantidadeBits(cap_memPrincipal, large_linha, cap_cache):
    totBits_Dados = cap_cache * large_linha * 1024
    linhas = base2Redutor(cap_cache / large_linha, 'kb')
    blocos = base2Redutor(cap_memPrincipal / large_linha, 'mb')
    totBits_Blocos = 2 ** linhas * blocos
    total = totBits_Dados + totBits_Blocos
    return total

print(f'''Para {quantidadeBits(16, 8, 32)}, são necessários:
{quantidadeBits(16, 8, 32) / 1024}KB''')    
```

5.6 Cálculo de formato de endereço para memórias cache com mapeamento associativo completo. 

Considere uma MP com 64MB de capacidade associada a uma memória cache que possui 2K linhas, cada uma com largura de 16 bytes. Determine o formato do endereço para ser interpretado pelo sistema de controle da cache. 


```python
def enderecoMemoria_associativoFull(cap_memPrincipal, linhas_qtd, large_linha):
    endereco = base2Redutor(cap_memPrincipal, 'mb')
    enderecoLinha = base2Redutor(linhas_qtd, 'k')
    larguraBloco = base2Redutor(large_linha, 'b')
    enderecoBloco = endereco - larguraBloco
    return [endereco - enderecoBloco, enderecoBloco]

vetor = enderecoMemoria_associativoFull(64, 2, 16)
print(f'''Byte: {vetor[0]} bits
Bloco: {vetor[1]} bits''')
```

5.7 Seja uma MP constituída de blocos com largura de 32 bytes, associada a uma cache com 64KB. Em dado instante o processador realiza um acesso, colocando o seguinte endereço (expresso em algarismos hexadecimais): 3FC92B6. Determine qual deverá ser o valor binário do campo bloco que será localizado pelo sistema de controle da cache. 

_Neste exemplo, utilizaremos o valor em forma decimal, reaproveitando o conversor de hexadecimal para binário em exemplos anteriormente ilustrados._

```python
def enderecamentoBloco(large_bloco, cap_Cache, hexa):
    enderecoBin_tamanho = Hex_Bin(hexa)
    byte_campo = base2Redutor(large_bloco, 'b')
    bloco_campo = enderecoBin_tamanho - byte_campo
    enderecoBin = dec_Bin(int(hexa, 16))
    
    # faz a correção dos zeros faltantes
    if len(enderecoBin) < enderecoBin_tamanho:
        while len(enderecoBin) < enderecoBin_tamanho:
            temp = enderecoBin[::-1]
            temp += '0'
            enderecoBin = temp[::-1]
    
    return enderecoBin[:bloco_campo]

print(f'Endereço do campo bloco: {enderecamentoBloco(32, 64, "3FC92B6")}')
```

5.8 Cálculo da quantidade de bits necessários para urna determinada memória cache, que funciona com mapeamento por conjunto de quatro. 

Considere um sistema de computação com uma memória cache de 32KB de capacidade, constituída de linhas com 8 bytes de largura e conjunto de 4. A MP possui uma capacidade de 16MB. 


```python
def bits_total(cap_Cahce, large_linha, conjunto, cap_memPrincipal):
    tot_BitsDados = cap_Cahce * large_linha * 1024
    enderecoLinhas = base2Redutor(cap_Cahce / large_linha, 'kb')
    enderecoConjuntos = base2Redutor((cap_Cahce / large_linha) / conjunto, 'k')
    enderecoBlocos = base2Redutor(cap_memPrincipal / large_linha, 'mb')
    CampoTag = enderecoBlocos - enderecoConjuntos
    tot_BitsTags = (2 ** enderecoLinhas / 1024) * (2 ** CampoTag / 1024) * 1024
    totalBits = tot_BitsDados + tot_BitsTags
    
    return int(totalBits)
    
print(f'''Para {bits_total(32, 8, 4, 16)}, serão necessários:
{int(bits_total(32, 8, 4, 16) / 1024)} Kbits''')
```


5.9 Cálculo de formato de endereço para memórias cache com mapeamento associativo por conjunto. 

Considere uma MP com 64MB de capacidade associada a uma memória cache que funciona com mapeamento associativo por conjunto de 4 e que possui 128KB, com linhas de largura de 16 bytes. Determine o formato do endereço para ser interpretado pelo sistema de controle da cache

```python
def mapeiaCampos(capacidade_memPrincipal, conjuntos, capCache, larguraLinha):
    endereco = base2Redutor(capacidade_memPrincipal, 'mb')
    larguraBloco = base2Redutor(larguraLinha, 'b')
    larguraConjunto = base2Redutor(conjuntos, 'b')
    
    #A partir daqui, há a redução das quantidades para a base 2
    enderecoBloco = base2Redutor(capacidade_memPrincipal / larguraLinha, 'm') 
    enderecoCache_linha = base2Redutor(capCache / larguraLinha, 'k')
    enderecoConjunto = enderecoCache_linha - larguraConjunto
    enderecoConjunto_Bloco = enderecoBloco - enderecoConjunto
    
    return [enderecoBloco - enderecoConjunto, enderecoConjunto, larguraBloco]

vet = mapeiaCampos(64, 4, 128, 16)

print(f'''A Tag será de {vet[0]} bits
O Conjunto terá {vet[1]} bits
E o bloco vai ter {vet[2]} bits''')
```


5.10 Seja uma MP constituída de blocos com largura de 32 bytes, associada a urna cache com 128KB; a cache usa mapeamento por conjunto de 4. Em dado instante o processador realiza um acesso, colocando o seguinte endereço (expresso em algarismos hexadecimais): 31C92B6. Determine qual deverá ser o valor binário do conjunto que será localizado pelo sistema de controle da cache. 

```python
def valorBin(large_bloco, cap_Cache, conjuntos, h):
    EnderecoBin = Hex_Bin(h)
    tamConj = base2Redutor(conjuntos, 'b')
    CampoByte = base2Redutor(large_bloco, 'b')
    qntLinhas = base2Redutor(cap_Cache / large_bloco, 'kb')
    conjunto = qntLinhas - tamConj
    tag = EnderecoBin - conjunto - CampoByte
    enderecoBinario= dec_Bin(int(h, 16))
    
    # corrige 0
    if len(enderecoBinario) < EnderecoBin:
        while len(enderecoBinario) < EnderecoBin:
            temp = enderecoBinario[::-1]
            temp += '0'
            enderecoBinario = temp[::-1]
    
    return enderecoBinario[tag:(tag + conjunto)]
    
    
print(f'Neste exemplo, o binário será: {valorBin(32, 128, 4, "31C92B6")}')
```
