```python
import random

mem_principal = []
registrador = ""
cont_instrução = 0
```

```python
#Função para a MP. Aqui, nos parametros, temos a quantidade de cédulas e o tamanho delas
def memoriaPrincipal(celulas: int, tamanho: int):
    for x in range(celulas):
        cell = []
        for y in range(tamanho):
            cell.append(0)
        mem_principal.append(cell)
```

```python
def controladora_memPrincipal(operador: False, addres: mem_principal, dado: 110):
    end_decimal = conversor_bin2dec(addres, 'Binário -> Decimal', 8)
    if (operador):
        if len(dado) < len(mem_principal[0]):
            dado_temporario = list(dado)
            dado_temporario.reverse()
            while len(dado_temporario) < len(mem_principal[0]):
                dado_temporario.append(0)
            mem_principal[end_decimal] = dado_temporario[::-1]
        elif len(dado) > len(mem_principal[0]):
            return print(f"ERROR #001 - Tamanho da palavra excede {len(mem_principal[0])} bits.")
        else:
            mem_principal[end_decimal] = list(dado)
            return print(f"Dado armazenado no índice {end_decimal} da memória com sucesso.")
    else:
        return ''.join(str(i) for i in mem_principal[end_decimal])
```

```python
def conversor_bin2dec(valor_entrada, type, size):
    valor_convertido = 0

    if type == "dec->bin":
        numero = int(valor_entrada)
        string = ''

        while numero > 0:
            rest = int(numero % 2)
            string += str(rest)
            numero = (numero - rest) / 2

        if(len(string) < size):
            lista_temporaria = list(string)

            while len(lista_temporaria) < size:
                lista_temporaria.append('0')

            valor_convertido = ''.join(lista_temporaria[::-1])

    if type == "Binário -> Decimal":
        resultado = 0
        stringToList = list(valor_entrada)

        for i in range(len(stringToList)):
            valor_temporario = int(stringToList[i])
            if(valor_temporario != 1 and valor_temporario != 0):
                return print("ERROR! O valor deve ser binário")
            else:
                resultado += valor_temporario * \
                    2 ** (len(stringToList) - i - 1)

        valor_convertido = int(resultado)

    return valor_convertido
```
